# Evaluación de Modelos para Detección de Fraude en Tarjetas de Crédito

**Este notebook tiene como objetivo evaluar varios modelos de Machine Learning entrenados para la detección de fraude en tarjetas de crédito. A continuación, se describen los pasos realizados en el notebook.**

## Importar Librerías y Funciones

**Primero, importamos las librerías necesarias y las funciones definidas en el archivo `evaluate.py` que se encuentra en la carpeta `src`.**


```python
# Evaluación.ipynb
import sys
sys.path.append('../src')

from evaluate import load_model, load_test_data, evaluate_model, plot_precision_recall_curve
```

## Nombres de los Modelos a Evaluar
**Definimos una lista con los nombres de los modelos que vamos a evaluar.**

```python
# Nombres de los modelos a evaluar
model_names = ["Logistic Regression", "Random Forest", "XGBoost", "LightGBM"]
```

## Evaluar los Modelos
**Para cada modelo en la lista, realizamos los siguientes pasos:**

- Cargamos el modelo entrenado utilizando la función `load_model`.
- Cargamos los datos de prueba utilizando la función `load_test_data`.
- Evaluamos el modelo utilizando la función `evaluate_model`.
- Graficamos la curva Precision-Recall utilizando la función `plot_precision_recall_curve`.

```python
for model_name in model_names:
    print(f"\n🔹 Evaluando modelo: {model_name}")
    model = load_model(model_name)
    X_test, y_test = load_test_data()
    y_pred, y_prob = evaluate_model(model, X_test, y_test)
    plot_precision_recall_curve(y_test, y_prob)
```
